## Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import pandas as pd

def reindex_with_custom_index(df):
    # Create a new index with values starting from 1 and incrementing by 2
    new_index = range(1, len(df) * 2, 2)
    
    # Use the reindex function to apply the new index
    df = df.reindex(new_index)
    
    # Reset the index to start from 1
    df.index = range(1, len(df) + 1)
    
    return df

# Example DataFrame
data = {'A': [10, 20, 30],
        'B': [40, 50, 60],
        'C': [70, 80, 90]}

df = pd.DataFrame(data)

# Call the function to re-index the DataFrame
df = reindex_with_custom_index(df)

print(df)
```

## Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console. For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should calculate and print the sum of the first three values, which is 60.

```python
import pandas as pd

def calculate_sum_of_first_three(df):
    # Use iloc to select the first three values in the 'Values' column
    first_three_values = df['Values'].iloc[:3]
    
    # Calculate the sum of the selected values
    sum_of_first_three = first_three_values.sum()
    
    # Print the result to the console
    print("Sum of the first three values:", sum_of_first_three)

# Example usage:
# Assuming you have a DataFrame df with a 'Values' column
# containing [10, 20, 30, 40, 50]
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
calculate_sum_of_first_three(df)
```

## Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
import pandas as pd

def add_word_count_column(df):
    # Apply a lambda function to each row in the 'Text' column to count words
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    
# Example usage:
# Assuming you have a DataFrame df with a 'Text' column
# containing text data
df = pd.DataFrame({'Text': ["My Name Is Rishikesh Kumar , I Am From Bihar ",
                            "Just I Completed My Graduatuion Exam With First Grade And Now I AM Doing Study Of Data Science At Pw Intitute. "]})

add_word_count_column(df)

# The 'Word_Count' column will now contain the word counts for each row
print(df)
```


## Q5. How are DataFrame.size() and DataFrame.shape() different?


### In Pandas, DataFrame.size and DataFrame.shape are two different attributes that provide different information about a DataFrame:

#### DataFrame.size:

* DataFrame.size returns the total number of elements (cells) in the DataFrame.
* It calculates the size by multiplying the number of rows by the number of columns.
* The result is a single integer representing the total number of elements in the DataFrame.

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
size = df.size
print(size)  # Output: 6 (3 rows * 2 columns = 6 elements)
```

#### DataFrame.shape:

* DataFrame.shape returns a tuple representing the dimensions of the DataFrame, where the first element is the number of rows, and the second element is the number of columns.
* It provides information about the structure of the DataFrame in terms of its dimensions.
Example:

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
shape = df.shape
print(shape)  # Output: (3, 2) (3 rows, 2 columns)
```

##### In summary, DataFrame.size gives you the total number of elements in the DataFrame, while DataFrame.shape provides a tuple with the number of rows and columns in the DataFrame. Depending on your needs, you might use one or both of these attributes to understand and work with your data in Pandas.


## Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address. The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username' column.

#### You can achieve this by using the apply() function along with a custom lambda function to extract the username from each email address and create a new 'Username' column. Here's a Python function that does this:

```python
d= {"Email" : ["username@domain.com","john.doe@example.com"] }
df = pd.DataFrame(d)
```

```python
df
```

```python
import pandas as pd

def extract_username(email):
    # Split the email address at the '@' symbol and take the first part
    username = email.split('@')[0]
    return username

# Sample data
d = {"Email": ["username@domain.com", "john.doe@example.com"]}
df = pd.DataFrame(d)

# Apply the extract_username function to create the 'Username' column
df['Username'] = df['Email'].apply(extract_username)

# Display the DataFrame
print(df)
```

#### This function uses the split('@') method to split each email address at the '@' symbol and then extracts the first part (the username) using [0]. The extracted usernames are stored in the new 'Username' column of the DataFrame.

## Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows. For example, if df contains the following values:

```python
import pandas as pd
data = {"A":[3,8,6,2,9],
        "B":[5,2,9,3,1],
        "C":[1,7,4,5,2]
}
df = pd.DataFrame(data)
print (df)
```

Your function should select the following rows:   A   B   C

1  8   2   7

4  9   1   2

The function should return a new DataFrame that contains only the selected rows.

```python
import pandas as pd

# Sample DataFrame
data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}

df = pd.DataFrame(data)

def filter_dataframe(df):
    # Use boolean indexing to select rows that meet the conditions
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_rows

# Call the function to get the selected rows
selected_df = filter_dataframe(df)

# Display the selected DataFrame
print(selected_df)
```

## Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.


```python
import pandas as pd

# Sample DataFrame with a 'Values' column
data = {'Values': [3, 8, 6, 2, 9]}
df = pd.DataFrame(data)

def calculate_statistics(df):
    # Calculate mean, median, and standard deviation
    mean = df['Values'].mean()
    median = df['Values'].median()
    std_dev = df['Values'].std()
    
    return mean, median, std_dev

# Call the function to calculate statistics
mean_value, median_value, std_deviation = calculate_statistics(df)

# Print the results
print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation)
```


## Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.


```python
import pandas as pd

# Sample DataFrame with 'Sales' and 'Date' columns
data = {'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05', '2023-09-06', '2023-09-07', '2023-09-08', '2023-09-09', '2023-09-10'],
        'Sales': [100, 120, 130, 110, 150, 140, 160, 170, 180, 190]}

df = pd.DataFrame(data)

def calculate_moving_average(df, window_size=7):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Sort DataFrame by 'Date'
    df = df.sort_values(by='Date')
    
    # Calculate the moving average
    df['MovingAverage'] = df['Sales'].rolling(window=window_size, min_periods=1).mean()
    
    return df

# Call the function to calculate the moving average
df = calculate_moving_average(df)

# Display the updated DataFrame
print(df)
```


## Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

For example, if df contains the following values:

         Date

0  2023-01-01

1  2023-01-02

2  2023-01-03

3  2023-01-04

4  2023-01-05

Your function should create the following DataFrame:


         Date    Weekday

0  2023-01-01    Sunday

1  2023-01-02     Monday

2  2023-01-03    Tuesday

3  2023-01-04    Wednesday

4  2023-01-05    Thursday

The function should return the modified DataFrame.

```python
import pandas as pd

# Sample DataFrame with a 'Date' column
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df = pd.DataFrame(data)

def add_weekday_column(df):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Create a new 'Weekday' column containing the weekday names
    df['Weekday'] = df['Date'].dt.strftime('%A')
    
    return df

# Call the function to add the 'Weekday' column
df = add_weekday_column(df)

# Display the updated DataFrame
print(df)
```

## Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.


```python
import pandas as pd

# Sample DataFrame with a 'Date' column containing timestamps
data = {'Date': ['2023-01-05 10:00:00', '2023-01-15 14:30:00','2023-01-17 09:56:23','2023-01-20 12:09:45', '2023-02-05 08:45:00', '2023-01-25 16:20:00']}
df = pd.DataFrame(data)

def select_rows_in_date_range(df, start_date, end_date):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Use boolean indexing to select rows within the date range
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return selected_rows

# Call the function to select rows within the date range
start_date = '2023-01-01'
end_date = '2023-01-31'
selected_df = select_rows_in_date_range(df, start_date, end_date)

# Display the selected DataFrame
print(selected_df)
```
